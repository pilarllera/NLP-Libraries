```python
text2 = 'Arthur Schopenhauer nació el 22 de febrero de 1788 en el seno de una acomodada familia de Danzig. El padre de Arthur, Heinrich Floris Schopenhauer, fue un próspero comerciante que inició a su hijo en el mundo de los negocios, haciéndole emprender largos viajes por Francia e Inglaterra. Su madre, Johanna Henriette Trosenier, fue una escritora que alcanzó cierta notoriedad al organizar soirées (veladas) literarias en la ciudad de Weimar'
```

```python
import es_core_news_lg
nlp = es_core_news_lg.load()
doc = nlp(text1)
```

```python
# Recordamos cuáles son algunos de los rasgos lingüísticos más importantes de los que extrae spaCy

for token in doc:
  print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop)
```

```python
from spacy import displacy
```

```python
# Y que tenemos a nuestra disposición un módulo especializado en las visualización de datos, por ejemplo de los datos obtenidos por el NER

displacy.render(doc, style='ent', jupyter=True)
```

```python
# Comprobamos ahora algunas de las funcionalidades que textacy añade sobre spaCy, por ejemplo la de extracción de n-gramas (esto es, de conjuntos formados por n tokens)

doc = textacy.make_spacy_doc(text1, lang=es_core_news_lg.load())
list(textacy.extract.ngrams(doc, 3, filter_stops=True, filter_punct=True, filter_nums=False))
```

```python
# Observamos otra interesante funcionalidad de textacy: la de tener almacenados patrones morfológicos

pattern = textacy.constants.POS_REGEX_PATTERNS['en']['NP']
```

```python
# Veamos, por ejemplo, el patrón de los sintagmas nominales:

pattern
```

```python
# Y comprobemos cómo es posible extraer n-gramas determinados por un determinado patrón morfológico

list(textacy.extract.pos_regex_matches(doc, pattern))
```

```python
# Pregunta 1:
# Genera un patrón de extracción de sintagmas nominales en español que contemple la posibilidad de que incluyan adverbios (por ejemplo en "lugares muy lejanos")
pattern = '<DET>? <NUM>* (<ADJ> <PUNCT>? <CONJ>?)* (<NOUN>|<PROPN> <PART>?)+ <ADV>? <ADJ>?'
texto = "En lugares muy lejanos viven unos animales muy extraños. Los niños rubios juegan al tenis."
doc = textacy.make_spacy_doc(texto, lang=es_core_news_lg.load())
list(textacy.extract.pos_regex_matches(doc, pattern))
```

```python
# Pregunta 1 (ampliación):
# Amplía la cobertura de tu patrón para que reconozca sintagmas como "personas muy poco inteligentes" o "personas muy poco inteligentes pero muy válidas"
pattern = '<DET>? <NUM>* (<ADJ> <PUNCT>? <CONJ>?)* (<NOUN>|<PROPN> <PART>?)+ (<ADV>* <ADJ>? (<CCONJ> <ADV>? <ADJ>?)*)*'
texto = "Las personas muy poco inteligentes me caen fatal.En lugares muy lejanos viven unos animales muy extraños. Los niños rubios juegan al tenis. Hay personas muy poco inteligentes pero muy válidas"
doc = textacy.make_spacy_doc(texto, lang=es_core_news_lg.load())
list(textacy.extract.pos_regex_matches(doc, pattern))
```

```python
# no me salía el patrón poniendo <CONJ> así que decidí imprimir cada token de la frase junto a su etiqueta POS para ver si no la estaba escribiendo bien
# efectivamente el resultado es que CONJ está escrito con doble C y por eso el patrón no me lo reconocía
texto = "Hay personas muy poco inteligentes pero muy válidas"
doc = nlp(texto)
for token in doc:
  print(token.text, token.pos_)
```

La función extract.semistructured_statements de Textacy se utiliza para hacer una tarea de extracción de información semiestructurada. Está diseñada para identificar declaraciones o afirmaciones en un texto que no sigan una estructura rígida pero que contengan información significativa.
Para realizar esto le tenemos que dar una indicación (cue) que indique con qué está relacionada la estructura que buscamos.

En este caso, la función busca declaraciones que contengan la palabra "Trump" cuando este se relacione con el verbo "indicar" en el texto.

Algunos posibles usos de esta función podría ser el análisis de reviews de clientes para analizar sus opiniones de determinados productos; la extracción de datos financieros; resúmenes de noticias para extraer declaraciones clave de algunas personas importantes o afirmaciones de artículos de noticias.

```python
pattern = '<DET>? <NUM>* (<ADJ> <PUNCT>? <CONJ>?)* (<NOUN>|<PROPN> <PART>?)+ <ADV>? <ADJ>?'
```
