```python
# Ejercicio 4: Genera una gráfica de barras orientada horizontalmente que incluya la información del número de alumnos suspensos de cada curso de la carrera:
# 15 en primero, 8 en segundo, 10 en tercero y 2 en cuarto.
# Dibuja el ancho de las barras con 0.5 puntos.
# Las barras deben ser de color verde.
# La gráfica debe tener el título "Número de suspensos por curso en el grado de Lingüística".
import matplotlib.pyplot as plt

# Datos
x = ["Primero", "Segundo", "Tercero", "Cuarto"]
y = [15, 8, 10, 2]

# Gráfico de barras
fig, ax = plt.subplots()
ax.barh(x, width = y, height = 0.5, color = "green")
ax.set_title('Número de suspensos por curso en el grado de Lingüística')
# plt.show()
```
